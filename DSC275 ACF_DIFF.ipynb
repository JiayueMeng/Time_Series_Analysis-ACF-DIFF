#### Question 4
##### a) Is there seasonality in this time series?						(2 pts)
<p></p>
Yes, there is seasonality in this time series. We can clearly see the seasonality from the time series plot.


```python
Beverages_df = pd.read_csv(r'C:\Users\Joyce\Desktop\DSC 275\Beverages_Shipment_2020(1).csv')
```

```python
Beverages_df.plot('Month','Dollars (in Millions)')
plt.title('Time Series Plot of Beverages Shipment')
plt.ylabel('Beverages Shipment (in Millions)')
plt.xlabel('Time')
plt.xticks(rotation = -30)
plt.show()
```

##### b) Find the sample autocorrelation function for this data set. 			(2 pts) (For Python, you can use the “plot_acf” function in “statmodels” module. Plot at least 25-30 lags)

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(Beverages_df['Dollars (in Millions)'], lags = 30)
plt.title(r'Beverages Shipment - ACF', fontsize = 16)
plt.ylabel('Autocorrelation', fontsize = 16)
plt.xlabel('Lag', fontsize = 16)
plt.show()
```

##### c) From the autocorrelation plot in (b), what is the seasonal period? 		(2 pts)
<p></p>
If there is significant seasonality, the autocorrelation plot should show spikes at lags equal to the period. Therefore, from the autocorrelation plot in (b), we know that the seasonal period is 12 lags (we can see it from lag=0 to lag=12).


#### Question 5
##### a) Plot the coal production data and the sample autocorrelation function. 		(2 pts)

```python
Coal_df = pd.read_csv(r'C:\Users\Joyce\Desktop\DSC 275\Coal_Production_US_2020.csv')
```

```python
plt.figure(figsize=(12,7))
plt.plot(Coal_df['Year'],Coal_df['Coal Production, Short Tons in Thousands'])
plt.title('Time Series Plot of Coal Production', fontsize = 16)
plt.ylabel('Coal Production, Short Tons in Thousands', fontsize = 16)
plt.xlabel('Time', fontsize = 16)
plt.show()
```

```python
plot_acf(Coal_df['Coal Production, Short Tons in Thousands'], lags = 30)
plt.title(r'Coal Production - ACF', fontsize = 16)
plt.ylabel('Autocorrelation', fontsize = 16)
plt.xlabel('Lag', fontsize = 16)
plt.show()
```

##### b) Is the time series stationary or non-stationary ? 					(2 pts)
<p></p>
The time series is non-stationary. Because from the time series plot of coal production, we can see that the time series exhibits a trend. Also, the mean and variance are not constant over time. From the plot Coal Production - ACF, we know that its autocorrelation decreases over time, which also means that there is a trend. So the time series is non-stationary.

##### c) Plot the first difference of the time series and the sample autocorrelation function of the first difference. 


```python
diff = Coal_df.diff().dropna()
plt.plot(diff['Coal Production, Short Tons in Thousands'])
plt.title('Plot of the First Difference of Coal Production', fontsize = 16)
plt.ylabel('First Difference', fontsize = 16)
plt.show()
```

```python
plot_acf(diff['Coal Production, Short Tons in Thousands'], lags = 10)
plt.title(r'First Difference of Coal Production - ACF', fontsize = 16)
plt.ylabel('Autocorrelation', fontsize = 16)
plt.xlabel('Lag', fontsize = 16)
plt.show()
```

##### d) What impact has differencing had on the time series? Comment with respect to presence or absence of stationarity								(2 pts)
<p></p>
The first difference of a time series is the series of changes from one period to the next. If Y_t refers to the value of the time series Y at period t, then the first difference of Y at period t is Y_t-Y_t-1.
<p></p>
Differencing helps to stabilize the variance and the mean of a time series by removing changes in the level of a time series, and therefore or reducing trend and seasonality.
<p></p>
In this question, differencing helps us remove the trend of this series and change the series from non-stationary to almost stationary. We can see from the Plot of the First Difference of Coal Production that the series after differencing is almost stationary. From the plot First Difference of Coal Production - ACF, we see that the trend of the series is removed and the series becomes almost stationary.
